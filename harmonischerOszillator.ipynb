```python
#force: kraft
#velocity: geschwindigkeit
#position: ort
```

```python
def get_forces(x):
    #Potential: V(x) = 0.5 k * x**2
    k = 3
    #Kraft: F(x) = -dV/dx
    #F(x) = -k * x
    return - k * x
```

```python
def update_position(x, v, dt):
    x_next = x + v * dt 
    return x_next
#def update_position(x, v, dt):
#    return  x + v * dt 
```

```python
def update_velocity(v, m, f, dt):
    a = f/m # Beschleunigung 
    v_next = v +  a * dt
    return v_next
```

```python
x_old = 1
#v_old = -1
v_old = 0
m = 2
dt = 0.01

x_list = []
v_list = []
f_list = []

x_list.append(x_old)
v_list.append(v_old)

f = get_forces(x_old)
f_list.append(f)

steps = int(10/dt)
#for i in range(int(5/dt)):
for i in range(steps):
    f = get_forces(x_old)
    #print(x_old, f)
    x_next = update_position(x_old, v_old, dt)
    x_list.append(x_next)
    v_next = update_velocity(v_old, m, f, dt)
    v_list.append(v_next)
    f_list.append(f)
    x_old = x_next
    v_old = v_next
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

#x = [1, 2, 3, 4, 5]
#y = [1, 2, 3, 4, 5]

time = np.arange(steps + 1)* dt
#time = np.linspace(0,steps)
#print(time)

plt.plot(time, x_list)
plt.show()
```

```python
plt.subplot(3, 1, 1)
plt.plot(time, x_list, 'x')
#plt.plot(time, x_list, 'x')
plt.xlabel("time / seconds")
plt.ylabel("amplitude / meters")





plt.subplot(3, 1, 2)
plt.plot(time, v_list)
plt.xlabel("time / seconds")
plt.ylabel("velocity / ms$^{-1}$")

plt.subplot(3, 1, 3)
plt.plot(time, f_list)
plt.xlabel("time / seconds")
plt.ylabel("force / Newton")

plt.show()
```
